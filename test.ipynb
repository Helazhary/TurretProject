```python
import cv2
```

```python
cam = cv2.VideoCapture(2)

while True:
    ret, frame = cam.read()
    if not ret:
        continue
    frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
   
    cv2.imshow("camView", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
